```python
import geopandas as gpd, matplotlib as mpl, matplotlib.backends.backend_pdf
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

world.name = world.name.astype(str)

world.loc[world.name=="Dem. Rep. Congo",'name'] = 'Democratic Republic of the Congo'
world.loc[world.name=="Tanzania",'name'] = 'United Republic of Tanzania'
world.rename(columns = {'name':'location_name'}, inplace=True)
```

```python
# make sure names align for merge
assert len([i for i in data_density.location_name.unique() if i not in world.location_name.unique()])==0

shapefiles = world.merge(data_density, on = 'location_name', how = 'right')
```

```python
# convert to logspace
shapefiles['log_nrows'] = np.log(shapefiles.nrows)
```

```python
cause_n_colors_dict = {
    'Dietary iron deficiency':6,
    'Vitamin A deficiency':5,
    'Zinc deficiency':5,
    'Neural tube defects':7
}
```

```python
def map_cause_input_data_density(cause, threshold, n_colors, top_tick):
    fig, ax = plt.subplots(figsize=(20,10))

    ax.set_aspect('equal')
    ax.set_title(f'GBD {cause} cause model: input data density', fontsize = 24)
    ax.set_axis_off() #remove box with lat/lons around map

    # plot country outlines
    base = world[world.continent.isin(['Africa','Asia'])].plot(ax=ax,
                                                               color = 'white',
                                                               edgecolor='gray')
    #subset to var of interest
    rf_df = shapefiles[shapefiles.Cause==cause]
    
    # create var for topcoding
    rf_df['plot_var'] = rf_df['log_nrows']
    
    # define the width of each colorbin; we want from our lowest to second-highest var to represent (n_colors-1) units
    step = (list(rf_df.plot_var.sort_values())[-2] - list(rf_df.plot_var.sort_values())[1])/(n_colors-1)
    
    # threshold our highest value to be one "step" above our second-highest value
    rf_df.loc[rf_df.plot_var > threshold,'plot_var'] = list(rf_df.plot_var.sort_values())[-2] + step*(1.00001)
    

    scale_max = rf_df.plot_var.max()
    scale_min = rf_df.plot_var.min()
    
    # color in countries for which we have counts
    cmap = plt.cm.get_cmap('copper_r', n_colors) #colorscale, discretized to n_colors bins
    norm = mpl.colors.Normalize(vmin=scale_min, vmax=scale_max) #normalize color scale
    me = rf_df.plot(ax=base,
                    column='plot_var', #our logscale, topcoded var
                    cmap=cmap,
                    edgecolor = 'gray')
    
    # add colorbar
    ticks = np.arange(scale_min,scale_max,step)
    cbar = plt.colorbar(mpl.cm.ScalarMappable(norm=norm, cmap=cmap), ax=me, ticks = ticks)
    # add tick labels, rounded to clean-looking numbers
    # add the '78,100' because India fell just above 78,000, and round_ticks rounds down 
    tick_labels = [str(rf_df.nrows.min())] + [add_commas(round_ticks(np.exp(i))) for i in ticks[1:-2]] + ['78,100',top_tick]
    cbar.ax.set_yticklabels(tick_labels)
    # print label to indicate scale is log
    cbar.set_label('Log scale: rows of input data', labelpad=-105, y=-.02, rotation=0)
    
save_path = '/ihme/homes/beatrixh/notebooks/viv_rsc/data_prep/outputs/plots/gbd_cause_ntd_input_data_density_map.pdf'
# save_path = '/ihme/homes/beatrixh/repos/scratch/test_delete.pdf'

pdf = matplotlib.backends.backend_pdf.PdfPages(save_path)
for cause in ['Neural tube defects']:
    fig =  map_cause_input_data_density(cause, threshold = 11.3, n_colors = 6, top_tick= '>100,000')
    pdf.savefig(fig)
pdf.close()
```
