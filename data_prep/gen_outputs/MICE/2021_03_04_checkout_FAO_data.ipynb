```python
import pandas as pd
```

```python
path = '/ihme/homes/beatrixh/vivarium_research_lsff/data_prep/inputs/FAOSTAT_food_supply_2018.csv'
```

```python
locs = pd.read_csv('/ihme/homes/beatrixh/vivarium_research_lsff/data_prep/inputs/lsff_countries.csv')
```

```python
locs.loc[locs.location_name=="Vietnam",'location_name'] = "Viet Nam"
```

```python
locs = locs[locs.location_name.notna()].location_name.tolist()
```

```python
df = pd.read_csv(path)
```

```python
df.loc[(df.Area=="China, mainland"),"Area"] = "China"
```

```python
areas = df[df.Area.isin(locs)].Area.unique()
```

```python
set(locs).difference(set(areas))
```

```python
grains  = ['Wheat and products', 'Rice and products', 'Maize and products']
```

```python
oils = [i for i in df.Item.unique() if 'Oil' in i]
```

```python
df = df[(df.Area.isin(areas)) & (df.Element=="Food supply (kcal/capita/day)") & (df.Item.isin(grains + oils))]
```

```python
df.Item = ['fao_' + i.lower().replace(",","").replace(" ","_") for i in df.Item]
```

```python
df = df.rename(columns={'Area':'location_name'})
```

```python
df = df[['location_name','Item','Value']].pivot(columns="Item", values = "Value", index = "location_name")
```

```python
save_path = '/ihme/homes/beatrixh/vivarium_research_lsff/data_prep/inputs/FAOSTAT_food_supply_2018.csv'
df.to_csv(save_path)
```
