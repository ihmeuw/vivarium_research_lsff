```python
estimates_path = '/ihme/homes/beatrixh/vivarium_research_lsff/data_prep/outputs/lsff_data_estimated_03_11_2021.csv'
est = pd.read_csv(estimates_path)
```

```python
est_val_map = {'A': 'percent of population eating vehicle',
 'B': 'percent of population eating industrially produced vehicle',
 'C': 'percent of population eating fortified vehicle',
 'D': 'percent of vehicle that is industrially produced',
 'E': 'percent of vehicle that is fortified',
 'F': 'percent of market covered by fortified product'}
```

```python
est.head()
```

```python
est.dropna()
```

```python
est.standard.unique()
```

```python
est = est[(est.B_estimate==1) | (est.C_estimate==1)]
```

```python
#keep estimates only; column A is all real data
est = pd.melt(est, id_vars = ['location_name','vehicle','nutrient','standard'], value_vars = ['B','C'],
       var_name = 'value_description', value_name = 'value_mean')
```

```python
est.value_description = est.value_description.map(est_val_map)
```

```python
est['is_estimate'] = 1
```

```python
df = df.append(est, sort = False)
```
