## add CIs

```python
new_data = pd.concat([new_wheat,new_oil,fortified_regular_oil])
```

```python
# from our (previously final) input sheet, select extracted rows of data (no regression or multiplicative results)
# then drop any rows duplicated for different nutrients
w_meta = input_data[input_data.estimation_status=='na'].drop(columns = 'nutrient').drop_duplicates()
```

```python
# use ratio of scale over mean to estimate scale for new industry datapoints
w_meta['scale_over_mean'] = (w_meta.value_975_percentile - w_meta.value_025_percentile) / w_meta.value_mean
r = w_meta.loc[(w_meta.scale_over_mean!=np.inf) & (w_meta.scale_over_mean!=0),['vehicle','scale_over_mean']]

r_mean = r.scale_over_mean.mean()
r.loc[r.vehicle=="maize flour",'scale_over_mean'] = r_mean

r = r.groupby('vehicle').mean().dropna().rename(columns={'scale_over_mean':'r'}).reset_index()
```

```python
vehicle_map = {'wheat flour':'industry wheat','oil':'industry oil'}
r.vehicle = r.vehicle.map(vehicle_map)
```

```python
r
```

```python
new_data = new_data.merge(r, on = 'vehicle', how = 'left')
```

```python
# input estimated CIs
new_data['lower'] = np.clip(new_data.value_mean - (new_data.r * new_data.value_mean)/2, 0, 100)
new_data['upper'] = np.clip(new_data.value_mean + (new_data.r * new_data.value_mean)/2, 0, 100)


new_data.loc[(new_data.value_025_percentile.isna()),'value_025_percentile'] = new_data.lower
new_data.loc[(new_data.value_975_percentile.isna()),'value_975_percentile'] = new_data.upper
```

```python
new_data = new_data.drop(columns=['lower','upper','r'])
```
