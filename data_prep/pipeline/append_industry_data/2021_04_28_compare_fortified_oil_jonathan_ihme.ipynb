```python
import pandas as pd, numpy as np
```

```python
final_save_path = '/ihme/homes/beatrixh/vivarium_research_lsff/data_prep/outputs/final_input_coverage_data_with_citations.csv'
ref_path = '/ihme/homes/beatrixh/vivarium_research_lsff/data_prep/outputs/lsff_input_coverage_data.csv'

df = pd.read_csv(final_save_path)
ref = pd.read_csv(ref_path)
```

```python
df.head()
```

```python
df.columns
```

```python
compare = df.merge(ref, on = ['location_id', 'location_name', 'vehicle',
       'value_description', 'nutrient'], how = 'outer', suffixes = ['_updated','_used'])
```

```python
compare.loc[
    (compare.value_mean_updated - compare.value_mean_used) > .1
][['location_name',
   'vehicle','value_description',
   'value_mean_updated',
   'value_mean_used','source_citation']].sort_values([
    'vehicle',
    'value_description',
    'location_name'
])
```

```python
compare.loc[
    (compare.vehicle=="oil") & 
    (compare.value_description=="percent of population eating fortified vehicle")
][['location_name','vehicle','value_description','value_mean_updated','value_mean_used','source_citation']]
```

```python
wmeta_save_path = '/ihme/homes/beatrixh/vivarium_research_lsff/data_prep/outputs/04_27_lsff_input_coverage_data_with_metadata.csv'
lit = pd.read_csv(wmeta_save_path)
```

```python
lit.vehicle.unique()
```

```python
lit_fortified_oil = lit.loc[
    (lit.vehicle=="oil") & 
    (lit.value_description=="percent of population eating fortified vehicle")
].sort_values('location_name')
```

```python
ind_fortified_oil = df.loc[
    (df.vehicle=="oil") &
    (df.value_description=="percent of population eating fortified vehicle")
].sort_values('location_name')
```

```python
lit_fortified_oil.columns
```

```python
comparison = lit_fortified_oil.merge(
    ind_fortified_oil,
    on = ['location_id', 'location_name', 'vehicle',
          'value_description', 'nutrient'],
    suffixes = ['_used','_requested']
)
```

```python
comparison = comparison[['location_id', 'location_name', 'sub_population_used', 'vehicle',
       'value_description', 'nutrient', 'value_mean_requested','value_mean_used',
       'value_025_percentile_used', 'value_975_percentile_used',
       'estimation_status_used',
       'source_citation_used', 'source_link_used', 'CI_source_used',
       'data_choice_notes_used', 'source_citation_requested']]
```

```python
comparison_save_path = '/ihme/homes/beatrixh/vivarium_research_lsff/data_prep/outputs/fortified_oil_requested_vs_used_4_28_2021.csv'

comparison.to_csv(comparison_save_path, index = False)
```

```python
comparison_save_path = '/ihme/homes/beatrixh/vivarium_research_lsff/data_prep/outputs/final_data_deliverables/oil_data_comparison.csv'
comparison.to_csv(comparison_save_path, index = False)
```
