```python
final_cols = ['location_id', 'location_name', 'sub_population', 'vehicle',
       'value_description', 'nutrient', 'value_mean', 'value_025_percentile',
       'value_975_percentile']
```

```python
subpop_map = {
    'total population': 'total population',
    'women of reproductive age': 'women of reproductive age',
    'under-5': 'under-5',
    '8-59 months': 'under-5',
    'children 12-36 months': 'under-5',
    'households with children <2 years old': 'NA',
    '0-24 months':'under-5',
    np.nan : 'NA'
}

df['subpop_val'] = df.sub_population.map(subpop_map)
```

```python
df.subpop_val.unique()
```

```python
df['nq_counts'] = df.groupby(['location_name','vehicle','value_description','nutrient']).transform('count').subpop_val
```

```python
df[(df.nq_counts > 1)].sort_values(['location_name','vehicle','value_description','nutrient']).set_index(['location_id','location_name','vehicle','value_description','nutrient'])[['nq_counts','subpop_val','value_mean','value_025_percentile','value_975_percentile']]
```

```python
assert(df.nq_counts.min()==1), "data formatting error"
```

```python
df['u5_applicable'] = False
df['wra_applicable'] = False

df.loc[(df.nq_counts==1),'u5_applicable'] = True
df.loc[(df.nq_counts==1),'wra_applicable'] = True
```

```python
df.loc[(df.nq_counts==2) &
       (df.subpop_val=='under-5'),
       'u5_applicable'] = True

df.loc[(df.nq_counts==2) &
       (df.subpop_val=='women of reproductive age'),
       'wra_applicable'] = True
```
